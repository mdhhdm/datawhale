# 压缩文件

```python
import numpy as np

outfile = r'.\task01.npz'
x = np.linspace(0, np.pi, 5)
y = np.sin(x)
z = np.cos(x)
np.savez(outfile, x, y, z_d=z)
data = np.load(outfile)
np.set_printoptions(suppress=True)
```

```python
print(data.files)  
```

```python
print(data['arr_0'])
```

```python
print(data['arr_1'])
```

```python
print(data['z_d'])
```

# 文本文件

```python
import numpy as np

outfile = r'.\task01.txt'
x = np.arange(0, 10).reshape(2, -1)
np.savetxt(outfile, x)
y = np.loadtxt(outfile)
print(y)
```

## 将numpy数组a中打印的项数限制为最多6个元素。

```python
import numpy as np
a = np.arange(10)
print(a)
```

```python
np.set_printoptions(threshold=6)
print(a)
```

## 打印完整的numpy数组a而不中断。

```python
import numpy as np
np.set_printoptions(threshold=20)
a=np.arange(50)
print(a)
```

```python
np.set_printoptions(threshold=np.iinfo(np.int).max)
print(a)
```
