```python
import numpy as np

x = np.arange(11,36).reshape(5,5)
print(x)
```

```python
np.amin(x)
```

```python
np.amin(x, axis=0)
```

```python
np.amin(x, axis=1)
```

# 计算方差

```python
x = np.arange(11,36).reshape(5,5)
```

```python
x
```

```python
np.var(x)
```

```python
np.mean((np.mean(x)-x)**2)
```

```python
np.var(x, ddof=1)
```

```python
np.sum((np.mean(x)-x)**2)/(x.size -1)
```

```python
np.var(x, axis=0)
```

```python
np.var(x, axis=1)
```

# 计算标准差

```python
np.std(x)
```

```python
np.sqrt(np.var(x))
```

```python
np.std(x, axis=0)
```

```python
np.std(x, axis=1)
```
